```python
# Check for mixed-type data in your df_cln dataframe

for col in df_cln.columns.tolist():
  weird = (df_cln[[col]].map(type) != df_cln[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cln[weird]) > 0:
    print (col)
```

```python
# melt the dataframe

df_melted = pd.melt(df_cln, id_vars=['Area', 'Months','Element'], var_name='Year', value_name='Temperature')
print(df_melted)
```

```python
#make sure Year is a number
df_melted['Year'] = df_melted['Year'].astype(int)
```

```python
df_melted['Element'] = df_melted['Element'].replace({
    'Temperature change': 'TC',
    'Standard Deviation': 'SD'})
df_melted.head()
```
